```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("credit_customers.csv")
```

```python
data.drop(['checking_status'], axis = 1 , inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
data.columns
```

### Here I have encoded all the columns for Logistic regression processiong, before transforming the Logistic regression all column value should be continuous(int or float)

```python
columns = ['credit_history', 'purpose','savings_status', 'employment','personal_status' ,'other_parties', 'property_magnitude', 'other_payment_plans', 'housing','job','own_telephone','foreign_worker','class'  ]
for i in columns:
    le = LabelEncoder()
    data[i] = le.fit_transform(data[i])
```

```python
data
```

```python
data.info()
```

```python
x = data.drop('class', axis = 1)
```

```python
y = data['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2)
```

```python
xtrain.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logR = LogisticRegression(solver = "newton-cg") # Here 4 methods are there as it is small data set so I am using newton-cg method
```

```python
logR.fit(xtrain, ytrain)
```

```python
ypred = logR.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ypred, ytest)
```
